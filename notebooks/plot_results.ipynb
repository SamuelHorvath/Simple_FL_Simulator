## Plotting

Displayed quantitites are averaged accross seeds with one standard error.

```python
%matplotlib inline

import sys
sys.path.insert(0, "../fl_sim/")
```

```python
from utils.plotting import plot

outputs_dir = '../outputs'  # where to look for the outputs of the experimets (default = '../outputs')
```

```python
class NameSpace(object):
    def __init__(self, adict):
        self.__dict__.update(adict)
```

```python
# Function to collect results from the experiments
def get_exps():
    exps = []
    for local_lr in local_lr_s:
        exp = {
            'identifier': identifier,
            'task': task,
            'agg': agg,
            'sampling': sampling,
            'local_opt': local_opt,
            'global_opt': global_opt,
            'global_lr': global_lr,
            'local_opt': local_opt,
            'local_lr': local_lr,
            'metric': metric,
            'train_metric': train_metric,
            'outputs_dir': outputs_dir,
        }
        exp = NameSpace(exp)
        exps.append(exp)
    return exps
```

Each experiment is saved based on the identifier, task name, global and local learning rates and seed. If you want to change the logic, update the `create_model_dir` function in `fl_sim/utils/utils.py`.

```python
identifier = "test"

task = 'femnist'
agg = 'mean'
sampling = 'uniform'
local_opt = 'sgd'
global_opt = 'sgd'
global_lr = 1.

local_lr_s = [0.1, 0.01, 0.001]


legend = [r'$\sf{FedAvg} (local-lr = 0.1)$', r'$\sf{FedAvg} (local-lr = 0.01)$', r'$\sf{FedAvg} (local-lr = 0.001)$']
```

```python
# criteria
train_metric = False # use train / test metric
metric = 'top_1_acc' # which metric to use for tuning

# what to plot 
kind = 'test_top_1_acc'
```

### Accuracy all

```python
exps = get_exps()

plot(
    exps, kind,
    log_scale=False,
    legend=legend,
    best=False, # best indicates whether to pick the best run among all hyperparameters (e.g., learning rates) all
    # file=f'test',  # if not commented, then saved as file to dir = '../plots/'
    x_label='comm. rounds', y_label=r'accuracy'
)
```

### Accuracy best

```python
legend = [r'$\sf{FedAvg}$']
```

```python
exps = get_exps()

plot(
    [exps[0]], kind,
    log_scale=False, legend=legend, best=True,
    # file=f'test',  # if not commented, then saved as file to dir = '../plots/'
    x_label='comm. rounds', y_label=r'accuracy'
)
```

### Loss all

```python
legend = [r'$\sf{FedAvg} (local-lr = 0.1)$', r'$\sf{FedAvg} (local-lr = 0.01)$', r'$\sf{FedAvg} (local-lr = 0.001)$']
```

```python
train_metric = True 
metric = 'loss'
 
kind = 'train_loss'
```

```python
exps = get_exps()

plot(
    exps, kind,
    log_scale=False, legend=legend, best=False,
    # file=f'test',  # if not commented, then saved as file to dir = '../plots/'
    x_label='comm. rounds', y_label=r'accuracy'
)
```

### Loss best

```python
legend = [r'$\sf{FedAvg}$']
```

```python
exps = get_exps()

plot(
    [exps[0]], kind,
    log_scale=False, legend=legend, best=True,
    # file=f'test',  # if not commented, then saved as file to dir = '../plots/'
    x_label='comm. rounds', y_label=r'accuracy'
)
```
